```python
flowers_images_dict={
    'roses':list(data_dir.glob('roses/*')),
    'daisy':list(data_dir.glob('daisy/*')),
    'dandelion':list(data_dir.glob('dandelion/*')),
     'sunflowers':list(data_dir.glob('sunflowers/*')),
    'tulips':list(data_dir.glob('tulips/*'))
}
```

```python
flowers_images_dict['roses']
```

```python
flowers_labels_dict ={
    'roses':0,
    'daisy':1,
    'dandelion':2,
    'sunflowers':3,
    'tulips':4
}
```

```python
img=cv2.imread(str(flowers_images_dict['roses'][0]))
```

```python
cv2.resize(img,(180,180)).shape
```

```python
X,y=[],[]
```

```python
for flower_name,images in flowers_images_dict.items():
    for image in images:
        img=cv2.imread(str(image))
        resized_img=cv2.resize(img,(224,224))
        X.append(resized_img)
        y.append(flowers_labels_dict[flower_name])
```

```python
X=np.array(X)
y=np.array(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)
```

```python
#this shows shower curtain in the image_labels available at the tensorflow hub.This dataset is to be downloaded for removing 
#above error .Now we need to add one last layer for the final prediction after freezing the other layers.
```

```python
feature_extraction_model="https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4"
```

```python
#just replaced 'classification by 'feature_vector' above expression.
```

```python
pretrained_model_without_top_layer= hub.KerasLayer(feature_extraction_model,input_shape=(224,224,3),trainable=False)
```

```python
#now comes the final model
```

```python
num_of_flowers=5
model=tf.keras.Sequential([
    pretrained_model_without_top_layer,
    tf.keras.layers.Dense(num_of_flowers)
])
```

```python
# now give the summary of the model telling parameteres used extra and when the pretrained model was built
```

```python
model.summary()
```

```python
model.compile(optimizer="adam",loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),metrics=['acc'])
```

```python
# REMEMBER "FROM LOGITS = TRUE" AND NOT "LOGITS=TRUE" SO WE NEED TO START USING FROM in above Line. 
```

```python
model.fit(X_train_scaled,y_train,epochs=5)
```

```python
model.evaluate(X_test_scaled,y_test)
```
